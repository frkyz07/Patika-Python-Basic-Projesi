# Patika Python Basic Projesi

```python
flat_list = [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
```

```python
def flatten(l): 
    l = flatten(l[0]) + (flatten(l[1:]) if len(l) > 1 else []) if type(l) is list else [l]
    return l
```

```python
flatten(flat_list)
```

```python
def reverse(list):
    list = list[::-1]
    return list
    
```

```python
a = [[1, 2], [3, 4], [5, 6, 7]]
reverse(a)
```
